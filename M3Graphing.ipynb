```python
import numpy as np
import pandas as pd
import random
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
f_in = open('C:/Users/bosco/Desktop/William & Mary/CTBA/M3/ConstructionTimeSeriesDatav2.csv')
ConstructTimeSeries = f_in.readlines()[1:]
f_in.close()
```

```python
for i in range(len(ConstructTimeSeries)): 
    ConstructTimeSeries[i]=ConstructTimeSeries[i].strip().split(',')
    ConstructTimeSeries[i]=int(ConstructTimeSeries[i][0]),str(ConstructTimeSeries[i][1]),int(ConstructTimeSeries[i][2]),int(ConstructTimeSeries[i][3]),int(ConstructTimeSeries[i][4])
    Month, Month_Year, Total_Construction, Private_Construction, Public_Construction = zip(*ConstructTimeSeries)
```

Time Series Plot of Total Construction shows that there are more construction occurred in month 40-85

```python
xLo = 30000.0
xHi = len(Private_Construction)
numPoints = 1000

fig, ax = plt.subplots()
ax.plot(Month,Total_Construction)
ax.plot(Month,Total_Construction,color='k',linewidth=1,label='Time Series Plot',marker='.')
ax.grid(False)
ax.xaxis.set_tick_params(which = 'both', top = True, bottom = True, labelbottom = True, direction = 'in')
ax.yaxis.set_tick_params(which = 'both', right = True, left = True, labelleft = True, direction = 'in')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_xlabel('Month',fontsize='20',fontname='Times New Roman')
ax.set_ylabel('Total Construction',fontsize='20',fontname='Times New Roman')
ax.set_xticks(Month[::5])
ax.set_xticklabels(Month[::146],fontsize='10')
ax.set_ylim(40000,max(Total_Construction))
ax.set_xlim(1,max(Month)+5)

ax.axvline(x=40,linewidth=1,color='k',label='low',linestyle='-')
ax.axvline(x=80,linewidth=1,color='k',label='High',linestyle='--')
ax.legend(loc=4,fontsize=20)
ax.axvspan(40,80,ymin=0,ymax=300,facecolor='b',alpha=0.1)
ax.text(60,50000,'Frequent Construction',fontsize=20,verticalalignment='center',horizontalalignment='center',fontname='Times New Roman')

fig.set_size_inches(30,30)
fig.suptitle("Time Series Plot of Total Construction",fontsize=70)
fig.savefig('M3Graphing.jpg',dpi=600)
```
